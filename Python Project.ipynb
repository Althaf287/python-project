```python
#Analysis Tasks:
```

```python
# Employee Distribution by Team
```

```python
team_distribution = df['Team'].value_counts()
print(team_distribution)
```

```python
team_percentage = (team_distribution / len(df)) * 100
```

```python
print(team_percentage)
```

```python
#Segregate Employees by Position
```

```python
position_distribution = df['Position'].value_counts()
print(position_distribution)
```

```python
#identify the predominant age group among employees.
```

```python
bins = [0, 18, 30, 40, 50, 100]
labels = ['0-17', '18-29', '30-39', '40-49', '50+']
```

```python
print(df.columns)
```

```python
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
print(df[['Age', 'age_group']].head())
```

```python
age_group_distribution = df['age_group'].value_counts()
```

```python
print(age_group_distribution)
```

```python
#Highest Salary Expenditure
```

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
```

```python
highest_team_salary = team_salary_expenditure.idxmax()
```

```python
position_salary_expenditure = df.groupby('Position')['Salary'].sum()
```

```python
highest_position_salary = position_salary_expenditure.idxmax()
```

```python
print(f"Team with highest salary expenditure: {highest_team_salary}")
print(f"Position with highest salary expenditure: {highest_position_salary}")
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df_clean = df.dropna(subset=['age', 'salary'])
```

```python
#Correlation Between Age and Salary
```

```python
correlation = df_clean['age'].corr(df_clean['salary'])
print(f"Correlation between age and salary: {correlation}")
```

```python
# Between Age and Salary
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='salary', data=df_clean)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
#Employee Distribution Across Teams
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values)
plt.title('Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
#Segregation by Position
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=position_distribution.index, y=position_distribution.values)
plt.title('Employee Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
#Age Group Distribution
```
